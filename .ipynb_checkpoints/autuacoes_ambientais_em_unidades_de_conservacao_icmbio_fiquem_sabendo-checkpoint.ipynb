```python
import pandas as pd
import numpy as np
```

```python
# importando a planilha
# mostrar o maximo de linhas 
df = pd.read_excel('eletronicos.xlsx')
pd.set_option('display.max_rows', 1000000000)
```

```python
# mostrar as areas com mais autuacoes
# df.pivot_table(
#     values='apreensao',
#     index='nome_uc',
#     aggfunc='count').sort_values(by='apreensao', ascending=False)

df.groupby(by='nome_uc').tabela_origem.value_counts().sort_values(ascending=False).head(5)
```

```python
# mostrar as areas com o valor mais alto de autuacoes somadas

df.groupby(by='nome_uc').valor_auto.sum().sort_values(ascending=False).head(5)
```

```python
# mostrar os valores medios de multa por autuacao em cada Unidade de Conservacao
# arredondar os centavos para duas casas


round((df.groupby(by='nome_uc').valor_auto.sum().sort_values(ascending=False) / df.groupby(by='nome_uc').tabela_origem.value_counts().sort_values(ascending=False)).sort_values(ascending=False), 2)
```

```python
# as descricoes mais decorrentes sobre as multas 
df.descricao.value_counts(ascending=False).head(5)
```

```python
# os tipos de infracao mais comuns 
df.tipo_infração.value_counts(ascending=False).head(5)
```

```python
# as multas mais altas
df.sort_values(by='valor_auto',ascending=False).head(5)
```

```python
# quando as multas aconteceram?

df.resample('M', on='data_lavratura').size().plot()
```

```python
# os estados mais atingidos
df.sigla_uf.value_counts(ascending=False)
```

```python
# as cidades/UF com mais multas
df_estados = df[['nome_municipio', 'sigla_uf']]

df_estados.value_counts(ascending=False)
```
